```python
import pandas as pd
from sklearn.model_selection import cross_val_score
```

```python
input_file = 'data/240-molecular-RdRp-data.csv'
data = pd.read_csv(input_file)
```

```python
data
```

```python
X = data.drop(columns=['best_affinity_kcal/mol', 'Drug', 'SMILES (pubchem)'])
y = data['best_affinity_kcal/mol']
```

```python
# setup PLS regression model
from sklearn.cross_decomposition import PLSRegression
pls = PLSRegression(n_components=10)
```

```python
# cross-validate the model
cv_scores = cross_val_score(pls, X, y, cv=5, scoring='neg_root_mean_squared_error')
print(f'Cross-validated RMSE scores: {-cv_scores.mean()}')
```
